```python
import torchxrayvision as xrv

dataset= xrv.datasets.SIIM_Pneumothorax_Dataset(imgpath='/mnt/nas/Users/Mori/zuzanna/siim-original/dicom-images-train/')
```

```python
import os 

# Access the Image IDs from the dataset
image_ids = dataset.csv['ImageId'].unique()  # Adjust if your dataset has a different way to access ImageIds

# Initialize a counter for existing image IDs
existing_count = 0

# Check for the existence of each ImageId
for img_id in image_ids:
    img_file = f"{img_id}"  # Adjust based on the actual naming convention
    img_path = os.path.join(dataset.imgpath, img_file)
    print(img_path)
    if os.path.isfile(img_path):
        existing_count += 1

# Calculate percentage of existing ImageIds
total_ids = len(image_ids)
percentage_existing = (existing_count / total_ids) * 100 if total_ids > 0 else 0

print(f"Total ImageIds: {total_ids}")
print(f"Existing ImageIds: {existing_count} ({percentage_existing:.2f}%)")
```

```python
import pandas as pd
import os

df = pd.read_csv('/home/zuzanna/MedKLIP/siim-pneumothorax-train-rle.csv.gz')
```

```python
file_map = {}
for root, directories, files in os.walk('/mnt/nas/Users/Mori/zuzanna/siim-original/dicom-images-train/', followlinks=False):
    for filename in files:
        filePath = os.path.join(root, filename)
        file_map[filename] = filePath
        
```

```python
file_map
```

```python
count = 0
for i in range(len(df)):
    # print(df['ImageId'][i] + '.dcm')
    if df['ImageId'][i] + '.dcm' in file_map:
        count += 1

count / len(df)
```

```python
df['exists'] = df['ImageId'].apply(lambda x: f"{x}.dcm" in file_map)

# Filter the DataFrame to only include rows where the file exists
df_existing = df[df['exists']].drop(columns='exists')
```

```python
df_existing.reset_index(drop=True, inplace=True)
```

```python
count = 0
for i in range(len(df_existing)):
    # print(df['ImageId'][i] + '.dcm')
    if df_existing['ImageId'][i] + '.dcm' in file_map:
        count += 1

count / len(df_existing)
```

```python
df_existing.to_csv('/home/zuzanna/MedKLIP/siim-pneumothorax-train-rle-existing.csv', index=False)
```
